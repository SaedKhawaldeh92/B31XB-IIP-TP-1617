```python
%matplotlib inline
%pprint off

from __future__ import print_function, division

import numpy as np
import matplotlib.pyplot as plt

from skimage import img_as_ubyte
from skimage.color import rgb2gray ,gray2rgb
from skimage.io import imread, imshow
from skimage.measure import label, regionprops
from skimage.morphology import binary_closing, binary_opening, disk, binary_erosion, binary_dilation
from skimage.filters import threshold_otsu
from skimage.transform import rescale
from skimage.draw import circle
```

## Segmentation of toy example

In this exercise, you will have to automatically find how many type of coins are present on the picture `coins.jpg`.
Therefore, read the image as `ubyte` and display it. In order to speed-up the subsequent processes, divide the scale of the image by a factor of 4.

```python
Path = './images/'
image_name = 'coins.jpg'
c_image = imread(Path.__add__(image_name))
c_image = img_as_ubyte(c_image)
c_image = rescale(c_image,0.25)
imshow(c_image)
plt.title('Rescaled Image ')
```

Subsquently, threshold the image using a global Otsu's thresholding after converting the image into gray scale. Display the image.

```python
c_image = rgb2gray(c_image)
thresh = threshold_otsu(c_image)
binary_image = c_image <= thresh
imshow(binary_image)
plt.title('Binary Image')
```

Clean the image using some morphological operations, i.e. erosion, dilation, opening, closing ...

```python
binary_erode_image = binary_erosion(binary_image,selem=disk(4))
plt.figure()
imshow(binary_erode_image)
plt.title('After Erosion')

binary_diltion_image = binary_dilation(binary_image,selem=disk(2))
plt.figure()
imshow(binary_diltion_image)
plt.title('After Dilation')

binary_open_image = binary_opening(binary_image)
plt.figure()
imshow(binary_open_image)
plt.title('After Opening')

binary_close_image = binary_closing(binary_image)
plt.figure()
imshow(binary_close_image)
plt.title('After Closing')
```

Using the functions [`label`](http://scikit-image.org/docs/stable/api/skimage.measure.html#skimage.measure.label) and [`regionprops`](http://scikit-image.org/docs/stable/api/skimage.measure.html#skimage.measure.regionprops) determine, the radius of each region and determine the number of type of coins in the image.

1. Show the image with the different labels.
1. How many regions do you find?
1. Display the radius of each region. Draw the corresponding circle on the image.

```python
from skimage.draw import circle_perimeter

image, number = label(binary_diltion_image,return_num=True)

print("The number of Detected Regions are= %d"%number )
plt.figure()
imshow(image)
imageprops = regionprops(image)

for s in range(number-1):
    if s == 0:
        continue
    radius = (imageprops[s].equivalent_diameter)/2.
    if radius > 30:
        area = imageprops[s].convex_area
        print("Radius= %d  " %radius)
        print("Area= %d  " %area)
        c_centroid = imageprops[s].centroid
        c_coord = circle_perimeter(int(c_centroid[0]),int(c_centroid[1]),int(radius))
        image[c_coord[0],c_coord[1]] = 255
    
plt.figure()
imshow(image)
```

Considering the techniques you used in the previous exercise, try to segment different objects in `objets4.jpg` image.

Using segmentation find the number of each object in the image. 

Considering that this is not as straight forward as the previous example, come up with your creative solutions and in case of failar explain the reasons and problems. 

```python
Path = './images/'
image_name = 'objets4.jpg'
c_image = imread(Path.__add__(image_name))
c_image = img_as_ubyte(c_image)
#c_image = rescale(c_image,0.25)
plt.figure()
imshow(c_image)
plt.title('Input Image ')

c_image = rgb2gray(c_image)
thresh = threshold_otsu(c_image)
binary_image = c_image <= thresh
plt.figure()
imshow(binary_image)
plt.title('Image after Threshold')
```

```python
binary_erode_image = binary_erosion(binary_image,selem=disk(2))
plt.figure()
imshow(binary_erode_image)
plt.title('After Erosion')

binary_diltion_image = binary_dilation(binary_image,selem=disk(8))
plt.figure()
imshow(binary_diltion_image)
plt.title('After Dilation')

binary_open_image = binary_opening(binary_image)
plt.figure()
imshow(binary_open_image)
plt.title('After Opening')

binary_close_image = binary_closing(binary_image)
plt.figure()
imshow(binary_close_image)
plt.title('After Closing')
```

```python
image, number = label(binary_diltion_image,return_num=True)

print("The number of Detected Regions are= %d"%number )

Imageprops = regionprops(image)
Numberofmarkers = 0
NumberofGums = 0

for s in range(number):
    area = Imageprops[s].convex_area
    if area > 9200: #classify them on the base of there area
        Numberofmarkers +=  1
    else: 
        NumberofGums += 1

print("Markers Count are : %d"%Numberofmarkers)
print("Gum Counts are : %d"%NumberofGums)
```
